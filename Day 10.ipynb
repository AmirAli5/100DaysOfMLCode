## Delay in AIDS Reporting in England and Wales prediction using SVM Regression

#### Part 1: Data Preprocessing

```python
# import Libraries
import pandas as pd                      # used for import the dataset
import numpy as np                       # used for multidimensional array
import matplotlib.pyplot as plt          # used for plot the Graph
```

```python
#import the dataset
dataset = pd.read_csv("aids.csv")
X = dataset.iloc[:,:-1].values       # predictor attribute 
y = dataset.iloc[:,-1].values        # target attribute
```

```python
# split the dataset into test set and train set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

#### Part 2: Building the Model

```python
# import SVM regression model from sscikit learn
from sklearn.svm import SVR
```

```python
# init the model
regressor = SVR(kernel = 'rbf')
```

```python
# fit the training model into our SVM classifer regression model
regressor.fit(X_train, y_train)
```

#### Part 3: Making a prediction and visualize the result

```python
# making a prediction
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Visualising the Regression results
plt.scatter(y_test, y_pred, color = 'red')
plt.plot(y_test, regressor.predict(X_test), color = 'blue')
plt.title('Delay in AIDS Reporting in England and Wales')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
