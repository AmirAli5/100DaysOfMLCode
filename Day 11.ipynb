## Recommendation System for Market Basket Optimisation using Apriori Algorithm (Association Rule)

```python
# import the libraries
import pandas as pd        # used for import the dataset
```

```python
# import the dataset
dataset=pd.read_csv('Market_Basket_Optimisation.csv', header=None)
transactions=[]
for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

#### Part 2: Building a model

```python
# import readymate apriori like sklearn
from apyori import apriori
rules= apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
```

#### Part 3: Making a Recommendation system

```python
# recommend system
result=list(rules)
result
```
