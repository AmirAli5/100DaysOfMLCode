## Mall Customers dataset prediction for different customer using k-Mean Clusterning

#### Part 1: Data Preprocessing

```python
# import the Libraries
import numpy as np        # used for multidimensional array
import pandas as pd       # used for import the dataset
import matplotlib.pyplot as plt   # used for ploting the Graph
```

```python
# import dataset
dataset=pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:, [3,4]].values
```

#### Part 2: using the elbow method to find optimal number of clusters

```python
# import kMean from sklean
from sklearn.cluster import KMeans
```

```python
# let's plot elbow method graph
wcss = []
for i in range(1,11):
    kmeans=KMeans(n_clusters= i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.xlabel('Number of Cluster')
plt.ylabel('wcss')
plt.show
```
