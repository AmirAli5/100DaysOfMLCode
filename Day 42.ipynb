## Fraud Detection Using Self Organization Map

#### Part 1: Data Preprocessing

```python
# import the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset = pd.read_csv('dataset\Credit_Card_applications.csv')
X = dataset.iloc [: ,:-1].values    # independent variables
y = dataset.iloc [: , -1].values    # dependent variables
```

```python
# feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
X = sc.fit_transform(X)
```

#### Part 2: Building The SOM Model

```python
# import the SOM model
from minisom import MiniSom
```

```python
# init the model
som = MiniSom( x = 10, y = 10, input_len = 15, sigma = 1.0, learning_rate = 0.5)
```

```python
# init the weight
som.random_weights_init(X)
```

```python
# traing the model
som.train_random(data = X, num_iteration = 100)
```

#### Part 3: Visualizing the result

```python
# making a self organization map
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o' , 's']
colors = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```
